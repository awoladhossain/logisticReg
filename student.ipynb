```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("student.csv")
dataset.drop(columns=["score"],inplace=True)
dataset.head(3)
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x="cgpa",y="placement",data=dataset)
plt.show()
```

```python
x = dataset["cgpa"]
y = dataset["placement"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
if x_train.ndim == 1:
    x_train = x_train.values.reshape(-1, 1)
if x_test.ndim == 1:
    x_test = x_test.values.reshape(-1, 1)
y_train = y_train.ravel()
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[3.2]])
```

```python
x_reshaped = x.values.reshape(-1, 1) if x.ndim == 1 else x

plt.figure(figsize=(5,4))
sns.scatterplot(x="cgpa",y="placement",data=dataset)
sns.lineplot(x="cgpa", y =lr.predict(x_reshaped),data=dataset, color="red")
plt.title("Placement vs CGPA")
plt.show()
```
