```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_set = pd.read_csv(r"Social_Network.csv")
data_set.drop(columns=["Salary"], inplace=True)
data_set.head(3)
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x="Age",y="Purchased",data=data_set)
plt.show()
```

```python
x = data_set["Age"]
y = data_set["Purchased"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
if x_train.ndim == 1:
    x_train = x_train.values.reshape(-1, 1)
if x_test.ndim == 1:
    x_test = x_test.values.reshape(-1, 1)
y_train = y_train.ravel()
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[33]])
```

```python
x_reshaped = x.values.reshape(-1, 1) if x.ndim == 1 else x

plt.figure(figsize=(5,4))
sns.scatterplot(x="Age",y="Purchased",data=data_set)
sns.lineplot(x="Age", y=lr.predict(x_reshaped), data=data_set, color="red")
plt.title('Logistic Regression Predictions')
plt.show()
```
